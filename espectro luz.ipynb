```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mediciones Dia 3/Espectro Monocromador/blancoposta_tratado.csv')
wavelenght = df[df.columns[0]].values
I = df[df.columns[1]].values
c = 299792458 #m/s
nu = (c/wavelenght)*1e-3
```

```python
fig, ax = plt.subplots(figsize = (12,7)) 

plt.rcParams['axes.autolimit_mode'] = 'round_numbers'
plt.rcParams['axes.xmargin'] = 0
plt.rcParams['axes.ymargin'] = 0

ax.plot(nu,I,label = f'Espectro de luz blanca')
ax.set_xlabel(r'$\nu $ [THz]',fontsize = 25)
ax.set_ylabel('Intensidad [mW]',fontsize = 25)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.yaxis.get_offset_text().set_fontsize('large') #Hace mas grande la letra que da la escala (el e-11)
ax.set_ylim(-0.020,0.25)
ax.set_xlim(min(nu),max(nu))
xlabels =np.linspace(min(nu),max(nu),10,endpoint = True)
ax.set_xticks(xlabels)
# print(xlabels)
ax.set_xticklabels([f'{number:.0f}' for number in xlabels])
ylbl = ax.get_yticks()
ax.axvline((c/454.64)*1e-3,ymax = max(I)/max(ylbl),color = 'tomato',label = rf'$\nu = ({(c/454.64)*1e-3:.0f}4 \pm 12)$ THz')

plt.legend(fontsize = 27)
# fig.savefig('Figuras/espectro blanco.png',bbox_inches='tight') #bbox tight para que no haya tando bordeblancoalpedo
```

```python
df = pd.read_csv('Mediciones Dia 3/Espectro Monocromador/420_tratado.csv')
wavelenght = df[df.columns[0]].values
I = df[df.columns[1]].values
I = I[1::]
c = 299792458 #m/s
nu = (c/wavelenght)*1e-3
nu = nu[1::]
```

```python
fig, ax = plt.subplots(figsize = (12,7)) 

ax.set_xmargin(0)
ax.autoscale_view() 

ax.plot(nu,I,label = rf'Espectro $\lambda = 420$')
ax.set_xlabel(r'$\nu$ [THz]',fontsize = 25)
ax.set_ylabel('Intensidad [mW]',fontsize = 25)
ax.tick_params(axis='both', which='major', labelsize=18)
ax.yaxis.get_offset_text().set_fontsize('large') #Hace mas grande la letra que da la escala (el e-11)
ax.set_ylim(-0.020,0.075)

#Seteo ejes
#Eje X
ax.set_xlim(min(nu),max(nu))
xlabels = np.linspace(min(nu),max(nu),10,endpoint = True)
ax.set_xticks(xlabels)
# print(xlabels)
ax.set_xticklabels([f'{number:.0f}' for number in xlabels])

#Eje Y
ax.set_ylim(-0.025,max(I))
ylabels = np.linspace(-0.025,max(I),5,endpoint = True)
ylabels = np.round(ylabels,3)
ax.set_yticks(ylabels)
ylbl = ax.get_yticks()



ax.axvline((c/428.84)*1e-3,ymax = 1,linewidth = 2,color = 'tomato',label = rf'$\nu = ({(c/428.84)*1e-3:.0f} \pm 12)$ THz')


plt.legend(fontsize = 25)
# fig.savefig('Figuras/informe/espectro azul.png',bbox_inches='tight') #bbox tight para que no haya tando bordeblancoalpedo
```
